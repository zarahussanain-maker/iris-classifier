# Iris Flower Classifier
A decision tree model to classify iris flower species.

```python
import os
import joblib
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

## Load Data

```python
iris = load_iris()
x = iris.data
y = iris.target
print(iris.feature_names, iris.target_names)
```

## Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Train Model

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
```

## Make Predictions

```python
y_pred = model.predict(x_test)
print("Predictions:", y_pred[:5])
print("True label:", y_test[:5])
```

## Evaluate Model

# Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Create outputs folder

```python
os.makedirs("outputs", exist_ok=True)
```

# Save model

```python
joblib.dump(model, "outputs/model.joblib")
```

# Save confusion matrix image

```python
plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()
plt.savefig("outputs/confusion_matrix.png")
```

## Model Results Explanation

The Decision Tree Classifier was trained using 80% of the dataset and tested on 20%.

The model achieved high accuracy, indicating strong performance in classifying iris flower species.

The confusion matrix shows that most predictions fall along the diagonal, meaning the model correctly classified the majority of samples.

This demonstrates that the model is effective for this classification task.
